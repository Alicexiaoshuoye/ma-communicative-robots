```python
#import packages
from csv import reader
import re 
import csv
```

## Test data without target sentences

```python
triple_0 = []
triple_1 = []
triple_2 = []
triple_3 = []
triple_4 = []
triple_5 = []
triple_6 = []

#load in csv file for test data 
with open('/Users/cayadehaas/PycharmProjects/Communicative_Robots/output_with_sentences.csv', 'r') as read_obj:
    # pass the file object to reader() to get the reader object
    dataset = reader(read_obj)
    for sentence in dataset: 
        if len(re.findall('&&', sentence[1])) == 0: 
              triple_0.append(sentence[1])
        if len(re.findall('&&', sentence[1])) == 1: 
               triple_1.append(sentence[1])
        if len(re.findall('&&', sentence[1])) == 2: 
               triple_2.append(sentence[1])
        if len(re.findall('&&', sentence[1])) == 3: 
               triple_3.append(sentence[1])
        if len(re.findall('&&', sentence[1])) == 4: 
               triple_4.append(sentence[1])
        if len(re.findall('&&', sentence[1])) == 5: 
               triple_5.append(sentence[1])
        if len(re.findall('&&', sentence[1])) == 6: 
               triple_6.append(sentence[1])
    print(triple_0)
    print(len(triple_2))
            
            
```

```python
#add triple groups to csv file
for i in range(0,7): 
    with open("triple_" + str(i+1) + '.csv', 'w', newline='') as myfile:
        wr = csv.writer(myfile)
        for sentence in locals()["triple_" + str(i)]: 
            wr.writerow([sentence])
```

## Test data with target sentences

```python
target_triple_1 = []
target_triple_2 = []
target_triple_3 = []
target_triple_4 = []
target_triple_5 = []
target_triple_6 = []
target_triple_7 = []

#load in csv file for test data 
with open('/Users/cayadehaas/PycharmProjects/Communicative_Robots/output_with_sentences.csv', 'r') as read_obj:
    # pass the file object to reader() to get the reader object
    dataset = reader(read_obj)
    for sentence in dataset: 
        if len(re.findall('&&', sentence[1])) == 0: 
               target_triple_1.append(sentence[-1])
        if len(re.findall('&&', sentence[1])) == 1: 
               target_triple_2.append(sentence[-1])
        if len(re.findall('&&', sentence[1])) == 2: 
               target_triple_3.append(sentence[-1])
        if len(re.findall('&&', sentence[1])) == 3: 
               target_triple_4.append(sentence[-1])
        if len(re.findall('&&', sentence[1])) == 4: 
               target_triple_5.append(sentence[-1])
        if len(re.findall('&&', sentence[1])) == 5: 
               target_triple_6.append(sentence[-1])
        if len(re.findall('&&', sentence[1])) == 6: 
               target_triple_7.append(sentence[-1])
    print(target_triple_2)
```

```python
#add triple groups to csv file
for i in range(1,8): 
    with open("target_triple_" + str(i) + '.csv', 'w', newline='') as myfile:
        wr = csv.writer(myfile)
        for sentence in locals()["target_triple_" + str(i)]: 
            wr.writerow([sentence])
```
